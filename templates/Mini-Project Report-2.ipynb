$$
\begin{array}{llll}
1.& R \xrightarrow{\alpha} 2R      & v_1 = [0,1],  & w_1 = \alpha y_2\\
2.& R+F\xrightarrow{\beta} 2F      & v_2 = [1,-1], & w_2 = \beta y_1y_2\\

$$
S =
\begin{bmatrix}
0 & 1 \\
1 & -1 \\
-1 & 0
\end{bmatrix}.
$$

```python
# Generate N random numbers from exponential distribution
def RandExp(lam,N):
    U =  np.random.rand(N)
    X = -1/lam*np.log(1-U)
    return X
```
